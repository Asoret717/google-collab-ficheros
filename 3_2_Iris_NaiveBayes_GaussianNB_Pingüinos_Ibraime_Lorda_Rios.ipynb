```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.naive_bayes import CategoricalNB
import numpy as np
import itertools

# Cargar el conjunto de datos de pingüinos
penguins = sns.load_dataset('penguins')

# Preprocesamiento
penguins.dropna(inplace=True)

# Codificar las columnas 'island' y 'sex' con LabelEncoder
for column in ['island', 'sex']:
    penguins[column] = LabelEncoder().fit_transform(penguins[column])

# Crear un LabelEncoder específico para 'species'
species_encoder = LabelEncoder()
penguins['species'] = species_encoder.fit_transform(penguins['species'])

# Dividir en entrenamiento y prueba
X = penguins.drop('species', axis=1)
y = penguins['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y ajustar CategoricalNB
cnb_classifier = CategoricalNB()
cnb_classifier.fit(X_train, y_train)
y_pred_cnb = cnb_classifier.predict(X_test)

# Calcular precisión
accuracy_cnb = accuracy_score(y_test, y_pred_cnb)

# Obtener matriz de confusión
cm = confusion_matrix(y_test, y_pred_cnb)

# Visualización de la matriz de confusión
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title(f'Accuracy: {accuracy_cnb:.2f}\nCategoricalNB\nNormalized confusion matrix')
plt.colorbar()
tick_marks = np.arange(len(species_encoder.classes_))
plt.xticks(tick_marks, species_encoder.classes_, rotation=45)
plt.yticks(tick_marks, species_encoder.classes_)

# Etiquetar los ejes
plt.xlabel('Predicted label')
plt.ylabel('True label')

# Mostrar los valores en la matriz
thresh = cm.max() / 2.
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, cm[i, j],
             horizontalalignment="center",
             color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```
